1. Data yang saya adalah informasi nilai mahasiswa dengan beberapa mata kuliah. Kolom yang mencakup terdapat Nama Mahasiswa, Nilai VDI, Nilai Komstat, Nilai AI, Rata-rata, dan Status Kelulusan

2. * Tujuan : Menampilkan distribusi jumlah mahasiswa berdasarkan Status Kelulusan

  * Pengguna : Dosen atau pihak akademik yang ingin memantau performa mahasiswa.

  * Fungsi : Laporan akademik atau dashboard pemantauan nilai.

  * Tone : Profesional dan informatif, dengan visualisasi yang jelas dan mudah dipahami.


```python
import pandas as pd

data = {
    'Nama Mahasiswa': ['Baruna', 'Virdio', 'Ahmed', 'Dhafin', 'Danang', 'Haikal', 'Bema', 'Dappa', 'Irvan', 'Ferdy'],
    'VDI': [85, 68, 92, 77, 60, 88, 74, 95, 70, 82],
    'Komstat': [90, 72, 80, 65, 85, 91, 75, 90, 65, 88],
    'AI': [78, 85, 88, 70, 60, 83, 79, 92, 75, 80]
}

df = pd.DataFrame(data)

df['Rata-rata'] = df[['VDI', 'Komstat', 'AI']].mean(axis=1)
df['Status Kelulusan'] = df['Rata-rata'].apply(lambda x: 'Lulus' if x >= 75 else 'Tidak Lulus')

print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pie chart untuk proporsi kelulusan mahasiswa
plt.figure(figsize=(7, 7))
df['Status Kelulusan'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])

# Menambahkan judul
plt.title('Proporsi Mahasiswa Lulus vs Tidak Lulus', fontsize=16)

# Menampilkan plot
plt.show()
```
